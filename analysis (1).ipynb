```python
df.info()
#extarcting info about our dataframe df
#as we can see there are various issues we need to handle
#firstly, our dataframe has missing values that are represented by some special character like '.','-,'_';
#secondly, instead of floating point datatype, the extracted data is in object format i.e. strings or character
#there are some columns that have lot of missing values
#features of our dataset are neither standarized nor normalized
```

```python
#dropping all such features with number of missing values>10
for cols in df.columns:
    if df[cols].isnull().sum()>10:
        df=df.drop([cols],axis=1)
```

```python

df.at[30,'railway route(kms)']='0'
```

```python
#Replacing all special characters present in our dataset (representing NULL values) to 0
for cols in df.columns:
    df[cols]=df[cols].replace('_','0')
    df[cols]=df[cols].replace('-','0')
    df[cols]=df[cols].replace('.','0')
```

```python
#converting object datatype columns or features to floating point number that is real numbers
for i in range(len(df.columns)):
    if i!=0:
        df[df.columns[i]]=df[df.columns[i]].astype(float)
```

```python
df.info()
#As we can see now ,we have converted our required features's object datatype to float
```

```python
#assigning a seperate dataframe for storing all statistics of numeris features of our dataset
df1=df.describe()
```

```python
#missing data are repesented by 0
#lets see how many missing data we have in each column
dict={}
for cols in df.columns:
    count=0
    for rows in df[cols]:
        if rows==0:
            count=count+1
    dict[cols]=count
```

```python
#Now all our missing data are represented by 0, Now to make our analysis we have to fill these missing data
#We will be fitting a regression model to find expected value of the missing data
#We will be using sklearn's linear regression
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
for i in range(len(df.columns)):
    if i!=0 and dict[df.columns[i]]>0:
        cols=df.columns[i]
        arr=[]
        arr1=[]
        for j in range(1,len(df[cols])):
            if df.at[j,cols]!=0:
                lis=[df.at[j,cols],df.at[j,"ease of doing business(rank)"]]
                arr.append(lis)
            else:
                arr1.append(df.at[j,"ease of doing business(rank)"])
        if(len(arr1)>0):
            arr2=np.array(arr)
            arr3=arr2[:,1].reshape(-1,1)
            arr4=np.array(arr1)
            arr5=arr4.reshape(-1,1)
            regr=LinearRegression()
            regr.fit(arr3,arr2[:,0])
    #         print(dict[cols])
    #         print(arr1)
            y=regr.predict(arr5)
            count=0;
            for j in range(len(df[cols])):
                try:
                    if df.at[j,cols]==0:
#                         print(df1[min][cols])
                        df.at[j,cols]=y[count]
                        if(df.at[j,cols]<0):
                            df.at[j,cols]=0
                        count=count+1
                except:
                    print(cols)
        
            
    
```

```python
#Using StandardScaler , standardizing our data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data)
```

```python
data_std=scaler.transform(data)
```

```python
data_std
```

```python
standard_data=pd.DataFrame(data_std,index=data.index,columns=data.columns)
```

```python
#data after standardization
standard_data
```

```python
#using mean to get easo of doing business score
mean_series=[]
for ind in standard_data.index:
    lis=list(standard_data.loc[ind,:])
    sum=0;
    for i in range(len(lis)):
        if(i<3):
            sum=sum+lis[i]
        else:
            sum=sum-lis[i]
    mean_series.append(sum)
```

```python
standard_data.insert(len(standard_data.columns),"Simple_Weighted_Mean",mean_series,True)
```

```python
standard_data
```

```python
Result1=standard_data.sort_values(by=["Simple_Weighted_Mean"],ascending=False)
```

```python
#using minmaxscaler to normalize our dataset
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
data_norm=min_max_scaler.fit_transform(data)
```

```python
data_norm
#our normalized data
```

```python
normalized_data=pd.DataFrame(data_norm,index=data.index,columns=data.columns)
```

```python
#getting standard deviation for each features
std_series=[]
for cols in normalized_data.columns:
    stddev=np.std(normalized_data[cols])
    std_series.append(stddev)
```

```python
std_series
```

```python
#calc ease of doing business score based on std dev weighted mean
mean_series2=[]
for ind in normalized_data.index:
    lis2=list(normalized_data.loc[ind,:])
    sum2=0;
    for i in range(len(lis2)):
        if(i<3):
            sum2=sum2+(lis2[i]*std_series[i])
        else:
            sum2=sum2-(lis2[i]*std_series[i])
    mean_series2.append(sum2)
```

```python
normalized_data.insert(len(normalized_data.columns),"Stddev_Weighted_Mean",mean_series2,True)
```

```python
Result2=normalized_data.sort_values(by=['Stddev_Weighted_Mean'],ascending=False)
```
